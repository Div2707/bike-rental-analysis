## Q4.what temp, atmep, humidity, windspeed is favourable for rental

```python
# Create scatter plots to visualize relationships
plt.scatter(df['temp'], df['cnt'], label='Temperature')
plt.scatter(df['atemp'], df['cnt'], label='Atmospheric Pressure')
plt.scatter(df['hum'], df['cnt'], label='Humidity')
plt.scatter(df['windspeed'], df['cnt'], label='Windspeed')
plt.xlabel('Factors')
plt.ylabel('Rental Count')
plt.legend()
plt.show()

# Calculate correlation coefficients
correlation_matrix = df[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
print(correlation_matrix)
```

```python


# Calculate average rental count for each factor
avg_rental_by_temp = df.groupby('temp')['cnt'].mean()
avg_rental_by_atmep = df.groupby('atemp')['cnt'].mean()
avg_rental_by_humidity = df.groupby('hum')['cnt'].mean()
avg_rental_by_windspeed = df.groupby('windspeed')['cnt'].mean()

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Average Rental Count by Factors')

# Plot for Temperature
axes[0, 0].scatter(avg_rental_by_temp.index, avg_rental_by_temp)
axes[0, 0].set_title('Temperature')
axes[0, 0].set_xlabel('Temperature')
axes[0, 0].set_ylabel('Average Rental Count')

# Plot for Atmospheric Pressure
axes[0, 1].scatter(avg_rental_by_atmep.index, avg_rental_by_atmep)
axes[0, 1].set_title('Atmospheric Pressure')
axes[0, 1].set_xlabel('Atmospheric Pressure')
axes[0, 1].set_ylabel('Average Rental Count')

# Plot for Humidity
axes[1, 0].scatter(avg_rental_by_humidity.index, avg_rental_by_humidity)
axes[1, 0].set_title('Humidity')
axes[1, 0].set_xlabel('Humidity')
axes[1, 0].set_ylabel('Average Rental Count')

# Plot for Windspeed
axes[1, 1].scatter(avg_rental_by_windspeed.index, avg_rental_by_windspeed)
axes[1, 1].set_title('Windspeed')
axes[1, 1].set_xlabel('Windspeed')
axes[1, 1].set_ylabel('Average Rental Count')

plt.tight_layout()
plt.show()
```

## Q6.whether working day or holiday which has higher number of rentals and write its reason.

```python

# Calculate total rental counts for working days and holidays
working_day_count = df[df['workingday'] == 1]['cnt'].sum()
holiday_count = df[df['holiday'] == 1]['cnt'].sum()
print("working_day_count =",working_day_count)
print("holiday_count =",holiday_count)

# Determine which has a higher number of rentals
higher_rentals = 'workingday' if working_day_count > holiday_count else 'holiday'

# Print the result and reason
print(f"The higher number of rentals is on {higher_rentals}.")
if working_day_count > holiday_count:
    print("Reason: People tend to rent more on working days due to commuting and regular routines.",working_day_count)
else:
    print("Reason: Holidays might attract more people to engage in outdoor activities, leading to higher rentals.",holiday_count)
print("This much more rentals on working day than holiday",working_day_count-holiday_count)
```

```python
# Function to convert hour to AM/PM format
def convert_to_am_pm(hour):
    if hour == 0:
        return '12 AM'
    elif 1 <= hour <= 11:
        return f'{hour} AM'
    elif hour == 12:
        return '12 PM'
    else:
        return f'{hour - 12} PM'

# Apply the function to create a new 'hour_am_pm' column
dh['hr'] = dh['hr'].apply(convert_to_am_pm)

# Print the DataFrame with the new column
dh
```

```python
Q1. which hour has the most rentals
```

```python
hourly_rental_counts = dh.groupby('hr')['cnt'].sum()

# Find the hour with the highest rental count
hour_with_most_rentals = hourly_rental_counts.idxmax()

# Print the result
print("There might be a surge in rentals due to the end of the workday. People could be using rentals for their evening commute, leisure activities, or social outings after completing their daily responsibilities. Additionally, the pleasant late afternoon weather might encourage outdoor activities, contributing to the higher rental demand at this hour.")
print(f"The hour with the most rentals is {hour_with_most_rentals}.")
```
