```python
# Function to convert hour to AM/PM format
def convert_to_am_pm(hour):
    if hour == 0:
        return '12 AM'
    elif 1 <= hour <= 11:
        return f'{hour} AM'
    elif hour == 12:
        return '12 PM'
    else:
        return f'{hour - 12} PM'

# Apply the function to create a new 'hour_am_pm' column
dh['hr'] = dh['hr'].apply(convert_to_am_pm)

# Print the DataFrame with the new column
dh
```
